# Analise usando Querys

```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados/employee.csv')
pd.set_option('max_columns', 300000)
dados.head()
```

```python
dados.shape
```

```python
# Detalhes por Departamento
dados.groupby(dados['Department']).agg('mean')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg('mean')
```

```python
dados[['Age','MonthlyIncome','Department']].groupby(dados['Department']).agg('mean')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg('median')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg('sum').style.background_gradient(cmap ='copper')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg('min')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg('max')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg('var')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg('std')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg('count')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg('nunique')
```

```python
dados.loc[dados.groupby(dados['Department'])['MonthlyIncome'].idxmax()][['EmployeeNumber','Department','MonthlyIncome']]
```

```python
dados.loc[dados.groupby(dados['Department'])['MonthlyIncome'].idxmin()][['EmployeeNumber','Department','MonthlyIncome']]
```

```python
dados[['MonthlyIncome','Department']].groupby(dados['Department']).agg('mean').plot(kind = 'bar');
```

```python
dados[['MonthlyIncome','Department']].groupby(dados['Department']).agg('mean').plot(kind = 'line');
```

```python
dados[['MonthlyIncome','Department']].groupby(dados['Department']).agg('mean').plot(kind = 'area');
```

#### USANDO PIVOT TABLE

```python
dados.pivot_table(values ='MonthlyIncome', index = 'Department', aggfunc = 'mean')
```

```python
dados.pivot_table(values ='MonthlyIncome', index = 'Department', aggfunc = 'mean').style.background_gradient(cmap ='copper')
```

```python
dados.pivot_table(values ='MonthlyIncome', index = 'Department', aggfunc = 'mean').plot(kind = 'line');
```

## Diferena

```python
dados[['Age','Department']].groupby(dados['Department']).agg('mean')
```

```python
dados.pivot_table(values ='Age', index = 'Department', aggfunc = 'mean')
```

```python
dados[['Age','Department']].groupby(dados['Department']).agg(['min','max'])
```

```python
dados.pivot_table(values ='Age', index = 'Department', aggfunc = ['min','max'])
```
